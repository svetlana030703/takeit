```python
from sklearn.feature_extraction.text import CountVectorizer, HashingVectorizer, TfidfVectorizer

tfidf_vectorizer = TfidfVectorizer()
X_tfidf_vectorizer = tfidf_vectorizer.fit_transform(df['Краткое\nсодержание'])
X_tfidf_vectorizer.toarray()
```

```python
from sklearn.manifold import TSNE
X_embedded_visual = TSNE(n_components = 2).fit_transform(X_tfidf_vectorizer.toarray())
X_embedded_visual
```

```python
import seaborn as sns

sns.scatterplot(x=X_embedded_visual[:, 0], y=X_embedded_visual[:, 1], c=[i for i, _ in enumerate(df.Адресат)])
```

```python
from collections import Counter

unique_adresses = df.Адресат.unique()
for item in unique_adresses:
  df_adressat = df[df.Адресат == item]

  texts = df_adressat['Краткое\nсодержание'].tolist()

  concatenated_texts = ' '.join(texts)

  counter = Counter(concatenated_texts)
  top_5_words = counter.most_common(5)
  print(item, top_5_words)
```
